# Creating the model

```python
from transformers import pipeline

sentiment_analysis = pipeline("sentiment-analysis")

def sentiment_analysis_app(text):
  result = sentiment_analysis(text)[0]
  sentiment = result['label']
  confidence = result['score']

  return f"Sentiment: {sentiment.capitalize()} (Confidence: {confidence: .2f})"

sentiment_analysis("Test")[0]['score']
```

#Creating the Interface

```python
import gradio as gr

iface = gr.Interface(
    fn = sentiment_analysis_app,
    title = "Sentiment Analysis App",
    description = "Demonstration of a machine learning model",
    inputs = "image", outputs = "image",
    examples = [['I am happy'], ['I am neither happy nor sad']])

iface.launch()
```
